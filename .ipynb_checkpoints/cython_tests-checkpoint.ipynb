### Load data

```python
def print_attrs(name, obj):
    print(name)
    for key, val in obj.attrs.items():
        print (f"    {key}: {val}")
chignolin_path = '../data/md_datasets/chignolin-unbiased/CLN025-0-protein-ALL.h5'
f = h5py.File(chignolin_path, 'r')
#    for key, val in f.attrs.items():
#        print (f"{key}: {val}")
#    f.visititems(print_attrs)
```

```python
traj = md.load_hdf5(chignolin_path)
pairs = list(combinations(traj.top.select("symbol != H"), 2))
distances = md.compute_distances(traj, pairs)
```

```python
start = 100000
stop = 150000
num_steps = 500
n_frames = len(traj)
stride = int((stop-start)/num_steps)
centers = distances[start:stop:stride,:]
test_pts = distances[::50,:]
```

```python
centers.shape
```

```python
l = 7
K_full = Matern(centers, nu=2.5, l=l)
plt.matshow(K_full)
plt.colorbar()
```

```python
n = K_full.shape[0] - 1
K = K_full[:-1, :-1] 
K_T = K_full[-1,:-1]
L = np.linalg.cholesky(K)
U = L.T
beta = np.linalg.solve(L, K_T)
Aop = pylops.MatrixMult(U)
```

```python
eps = 0.01
maxit = 5000
res = pylops.optimization.sparsity.FISTA(
    Aop, beta, maxit, eps=eps, tol=1e-3, returninfo=True
)
alpha_ista = res[0]
alpha_l2 = np.linalg.solve(K, K_T)
S = companion_matrix(alpha_ista)
```

```python
plt.plot(alpha_l2, 'r.')
plt.plot(alpha_ista, 'k-')
```

```python
#This should be replaced with fast version
vals, R_vecs = np.linalg.eig(S)
#logvals = np.log(vals)
```

```python
L_vecs = np.vander(vals, increasing=True)
Id = L_vecs@R_vecs
zero_mat = Id - np.diag(np.diag(Id))
np.abs(zero_mat).max()
```

```python
mask = (np.imag(vals) >= 1e-8)
figure, axes = plt.subplots()
axes.set_aspect(1)
axes.plot(np.real(vals), np.imag(vals), 'k.')
axes.set_xlabel(r'$\rm{Re}\, \lambda$')
axes.set_ylabel(r'$\rm{Im}\, \lambda$')
```

```python
sort_perm = np.argsort(np.abs(vals[mask]))
CVs = (Matern(test_pts, centers[:-1], l=l, nu=2.5)@R_vecs[:, mask][:, sort_perm])*vals[mask][sort_perm]
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
pca_CVs = pca.fit_transform(np.abs(CVs))
```

```python
1 - np.sum(pca.explained_variance_ratio_)
```

```python
def f(x):
    kde = gaussian_kde(pca_CVs[:,x])
    bounds = (pca_CVs[:,x].min(),pca_CVs[:,x].max())
    padding = 0
    cv_sample = np.linspace(bounds[0] - padding, bounds[1] + padding, 100)
    #print(kde.factor)
    plt.plot(cv_sample, -kde.logpdf(cv_sample) )
    plt.margins(0)
widgets.interact(f, x=widgets.IntSlider(min=0, max=pca_CVs.shape[1] - 1, step=1, value=0));
```
